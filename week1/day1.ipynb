```python
# imports
#Connecting to the LLM running locally

import os
from openai import OpenAI
from dotenv import load_dotenv


# Load variables from .env
load_dotenv()
# Read the API key
api_key = os.getenv("OPENAI_API_KEY")
print(api_key)
base_url = os.getenv("OPENAI_API_BASE_URL")
print(base_url)

client = OpenAI(
    base_url=base_url,
    api_key=api_key     # Any non-empty string works
)

response = client.chat.completions.create(
    model="gpt-5-nano",
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant."
        },
        {
            "role": "user",
            "content": "Explain why prompt engineering is important."
        }
    ]
)

print(response.choices[0].message.content)

# If you get an error running this cell, then please head over to the troubleshooting notebook!
```

```python
#Connecting to the LLM running on Google
#https://aistudio.google.com/api-keys
from dotenv import load_dotenv
import os
from google import genai
from IPython.display import Markdown

load_dotenv()
api_key = os.getenv("GOOGLE_API_KEY")
print("Google AI api key:", api_key)
client = genai.Client(
    api_key=os.getenv("GOOGLE_API_KEY")
)

for model in client.models.list():
    print(model.name)
response = client.models.generate_content(
    model="gemma-4-31b-it",
    contents="Explain Context Engineering in AI."
)

print(response.text)
```
